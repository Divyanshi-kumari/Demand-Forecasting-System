```python
from flask import Flask, render_template, request, jsonify
import random

app = Flask(__name__)

@app.route('/')
def home():
    return render_template('index.html')

@app.route('/predict', methods=['POST'])
def predict():
    data = request.get_json()
    period = data.get('period', 'weekly')
    sector = data.get('sector', 'North')

    if period == 'weekly':
        labels = [f"Week {i+1}" for i in range(12)]
    elif period == 'monthly':
        labels = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
    else:
        labels = ["Q1", "Q2", "Q3", "Q4"]

    base = {
        "North": 5000,
        "East": 6500,
        "West": 7000,
        "South": 6000,
        "Overall": 20000
    }.get(sector, 5000)

    actuals = [base + random.randint(-1500, 2000) for _ in labels]
    predictions = [x + random.randint(-1000, 1200) for x in actuals]

    return jsonify({
        "labels": labels,
        "actuals": actuals,
        "predictions": predictions
    })

if __name__ == '__main__':
    app.run(debug=False, use_reloader=False)
```
